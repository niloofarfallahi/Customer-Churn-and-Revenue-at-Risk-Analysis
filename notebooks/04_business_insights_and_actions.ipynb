```python
import pandas as pd
import numpy as np

scores = pd.read_csv("../data/processed/churn_scores_test_pipeline.csv")
coefs  = pd.read_csv("../data/processed/logreg_coefficients_pipeline.csv")
metrics = pd.read_csv("../data/processed/logreg_pipeline_metrics.csv")

scores.head(), coefs.head(), metrics
```

```python
def risk_tier(p):
    if p >= 0.70:
        return "High"
    if p >= 0.40:
        return "Medium"
    return "Low"

scores["Risk_Tier"] = scores["Churn_Prob"].apply(risk_tier)
scores["Risk_Tier"].value_counts(normalize=True)
```

```python
scores.groupby("Risk_Tier")["Actual_Churn"].mean().sort_values(ascending=False)
```

```python
top_drivers = coefs.head(15).copy()
top_drivers["Direction"] = np.where(top_drivers["Coefficient"] > 0, "Increases churn", "Reduces churn")
top_drivers
```
